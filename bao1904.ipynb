```python
import folium
import networkx as nx
import random
from geopy.distance import geodesic

G = nx.Graph()

nodes = {
    'UEH_CoSoA': (10.7818, 106.6946),
    'NgaTu_PhatDiem': (10.7800, 106.6980),
    'NgaTu_MacDinhChi': (10.7840, 106.6960),
    'VongXoay_DanChu': (10.7780, 106.6880),
    'Cho_BenThanh': (10.7724, 106.6981)
}

for node, coord in nodes.items():
    G.add_node(node, pos=coord)

edges = [
    ('UEH_CoSoA', 'NgaTu_PhatDiem', 500),
    ('UEH_CoSoA', 'NgaTu_MacDinhChi', 800),
    ('UEH_CoSoA', 'VongXoay_DanChu', 600),
    ('NgaTu_PhatDiem', 'Cho_BenThanh', 400),
    ('NgaTu_MacDinhChi', 'Cho_BenThanh', 900),
    ('VongXoay_DanChu', 'Cho_BenThanh', 1000),
    ('NgaTu_MacDinhChi', 'NgaTu_PhatDiem', 300)
]

random.seed(42)

for u, v, cap in edges:
    vol = random.randint(300, 1200)
    risk_index = vol / cap

    he_so_phat = 10 if risk_index > 1.5 else 1
    real_distance = geodesic(nodes[u], nodes[v]).meters
    routing_weight = real_distance * he_so_phat

    G.add_edge(u, v, risk=risk_index, weight=routing_weight)

best_path = nx.shortest_path(G, 'UEH_CoSoA', 'Cho_BenThanh', weight='weight')

m = folium.Map(location=[10.778, 106.695], zoom_start=15)

for u, v, data in G.edges(data=True):
    is_congested = data['risk'] > 1.5
    color = 'red' if is_congested else 'gray'
    line_weight = 6 if is_congested else 3

    folium.PolyLine(
        [nodes[u], nodes[v]],
        color=color,
        weight=line_weight,
        opacity=0.5,
        tooltip=f"Mức độ kẹt xe: {data['risk']:.1f}"
    ).add_to(m)

path_coords = [nodes[n] for n in best_path]
folium.PolyLine(
    path_coords,
    color='blue',
    weight=5,
    tooltip='Tuyến đường thay thế an toàn'
).add_to(m)

for n in best_path:
    folium.Marker(
        nodes[n],
        popup=n,
        icon=folium.Icon(color='green', icon='ok-sign')
    ).add_to(m)

m
```

```python
import folium
import random
from geopy.distance import geodesic

kho_hang = [
    {"id": "Kho_A", "coord": (10.7818, 106.6946)},
    {"id": "Kho_B", "coord": (10.7468, 106.6614)}
]

diem_giao = [
    (10.7724, 106.6981), (10.7865, 106.6925), (10.7333, 106.7000),
    (10.7500, 106.6500), (10.7600, 106.6700), (10.7900, 106.7100),
    (10.7400, 106.6800), (10.7800, 106.7050)
]

def tinh_tong_kc(route):
    total = 0
    for i in range(len(route) - 1):
        total += geodesic(route[i], route[i+1]).km
    return total

phan_cum = {k['id']: [k['coord']] for k in kho_hang}
for p in diem_giao:
    kho_gan_nhat = min(kho_hang, key=lambda k: geodesic(p, k['coord']).km)
    phan_cum[kho_gan_nhat['id']].append(p)

tuyen_toi_uu = {}
for kid, points in phan_cum.items():
    start = points[0]
    unvisited = points[1:]
    route = [start]
    current = start
    while unvisited:
        next_p = min(unvisited, key=lambda p: geodesic(current, p).km)
        route.append(next_p)
        unvisited.remove(next_p)
        current = next_p
    route.append(start)
    tuyen_toi_uu[kid] = route

tuyen_khong_toi_uu = []
temp_points = diem_giao.copy()
random.shuffle(temp_points)
tuyen_khong_toi_uu = [kho_hang[0]['coord']] + temp_points + [kho_hang[0]['coord']]

kc_toi_uu = sum(tinh_tong_kc(r) for r in tuyen_toi_uu.values())
kc_goc = tinh_tong_kc(tuyen_khong_toi_uu)

print(f"Quãng đường không tối ưu: {kc_goc:.2f} km")
print(f"Quãng đường tối ưu (Multi-depot): {kc_toi_uu:.2f} km")
print(f"Hiệu quả tiết kiệm: {((kc_goc - kc_toi_uu) / kc_goc * 100):.1f}%")

m = folium.Map(location=[10.76, 106.68], zoom_start=13)

for k in kho_hang:
    folium.Marker(k['coord'], popup=k['id'], icon=folium.Icon(color='red', icon='home')).add_to(m)

for p in diem_giao:
    folium.Marker(p, icon=folium.Icon(color='blue', icon='info-sign')).add_to(m)

colors = ['green', 'orange']
for i, (kid, route) in enumerate(tuyen_toi_uu.items()):
    folium.PolyLine(route, color=colors[i], weight=5, opacity=0.8, tooltip=f"Tuyến {kid}").add_to(m)

m
```

```python
import folium
import numpy as np
from sklearn.cluster import KMeans

np.random.seed(42)
don_hang = np.random.normal(loc=[10.7769, 106.7009], scale=0.03, size=(500, 2))

so_luong_kho = 3
kmeans = KMeans(n_clusters=so_luong_kho, random_state=42, n_init=10)
kmeans.fit(don_hang)

vi_tri_kho = kmeans.cluster_centers_
nhan_cum = kmeans.labels_

m = folium.Map(location=[10.7769, 106.7009], zoom_start=12, tiles='cartodbpositron')

mau_sac = ['blue', 'green', 'purple']

for i in range(len(don_hang)):
    folium.CircleMarker(
        location=don_hang[i],
        radius=2,
        color=mau_sac[nhan_cum[i]],
        fill=True,
        opacity=0.4
    ).add_to(m)

for i, kho in enumerate(vi_tri_kho):
    folium.Marker(
        location=kho,
        popup=f"Vị trí MFC {i+1} tối ưu",
        icon=folium.Icon(color='red', icon='star')
    ).add_to(m)

    folium.Circle(
        location=kho,
        radius=3000,
        color='red',
        fill=True,
        fill_opacity=0.1,
        popup="Vùng phủ sóng 15 phút (3km)"
    ).add_to(m)

m
```
